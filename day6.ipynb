```python
print("Hello ram, Welcome to system")
print("Hello sita, Welcome to system")
print("Hello hari, Welcome to system")
print("Hello osin, Welcome to system")

# function define
def student(name):      #name = parameter
    return f"Hello, {name}! Welcome to system"

# function call
print("-----------------Names-----------------")
print(student("sita"))      # sita = argument
print(student("ram"))
print(student("hari"))
```

```python
price = 500
discount_amount = 400

discount_percentage = discount_amount / price * 100

def calculate_discount(price, discount_percent):
    print("The price of monitor:", price)
    print("The discount percentage:", discount_percent, "%")

# function call 
calculate_discount(price, discount_percentage)
```

```python
c = 80
f = 80

def c_to_f(c):
    return (c * 9/5) + 32

def f_to_c(f):
    return (f - 32) * 5/9

# Function call and print
print("Celsius to Fahrenheit:", c_to_f(c))
print("Fahrenheit to Celsius:", f_to_c(f))
```
