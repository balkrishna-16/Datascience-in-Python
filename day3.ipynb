```python
n = 5

for i in range(1, n+1):
    for j in range(n-i):
        print(" ", end="")
    for k in range(2*i-1):
        if k == 0 or k == 2*i-2 or i == n:
            print("*", end="")
        else:
            print(" ", end="")
    print()
```

```python
marks = [
    int(input("Enter mark of English: ")),
    int(input("Enter mark of Math: ")),
    int(input("Enter mark of Science: "))
]

total = sum(marks)
per = total / len(marks)

if per >= 80:
    grade = "A"
elif per >= 60:
    grade = "B"
else:
    grade = "C"

print(f"You passed with {per:.2f}% in grade {grade}")
```
