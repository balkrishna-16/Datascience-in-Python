```python
def login_system(user_role):
    print("Checking login creditials...")

    #Nested function: only accessible after login check
    def access_dashboard():
        print("Welcome to Dashboard")

    #Permission Check
    if user_role == "admin":
        print("Login succesful ")
        access_dashboard()     #inner function runs only if allowed

    else:
        print("Access denied ")

login_system("admin")
login_system("guest")
```

```python
def greet(name):
    return"Hello" + name
# Higher Order function
def process(func, value):
    return func(value)
print(process(greet," Sandesh"))
```

#Lambda

```python
square = lambda x: x*x         #x1: x1*x1
square(5)
```

```python
#Add two number using lamada
add = lambda x,y:x+y
add(3,2)
```

```python
#Use lamada to check if a number is even
is_even = lambda x: x % 2 == 0
is_odd = lambda x: x % 2 != 0

print(is_even(4))  # True
print(is_even(7))  # False
```

```python
#Get maximum of two numbers.
x_greater = lambda x, y : x>y 
y_greater = lambda x,y : x<y

print(x_greater(7,5))
print(y_greater(10,8))
```

```python
# minimum of two nums
x_lower = lambda x, y : x<y 
y_lower = lambda x,y : x>y

print(x_lower(7,5))
print(y_lower(10,8))
```

```python
#Get maximum of three numbers.
x_greater = lambda x, y,z : x>y |x>z
y_greater = lambda x,y,z : y>x  |y>z
z_greater = lambda x,y,z : z>y  |z >x


print(x_greater(11,5,9))
print(y_greater(7,8,9))
print(z_greater(10,8,9))
```

```python
max_num = lambda x, y, z: x if (x >= y and x >= z) else (y if y >= z else z)

print(max_num(11, 5, 9))
print(max_num(10, 8, 6))
print(max_num(4, 15, 6))
```

```python
#Return lenght of a string using lamada
x_length = lambda s: len(s)

print(x_length("hello"))  
print(x_length("Balkrishna"))  
```

```python
#map() applies a function to every item in a list
# Syntax:
# map(function, iterable)
num = [1,2,3,4]
result = list(map(lambda x: x, num))
print(num)
```

```python
#Conver list of strings to uppercase
words = ["Hello", "World", "Python"]
result = list(map(lambda u: u.upper(),words))
print(result)

#Lowercase from list of string
words = ["Hello", "World", "Python"]
result = list(map(lambda l: l.lower(),words))
print(result)
```

```python
#Add 10 to Each element in list
Nums = [1,2,3,4,5]
result = list(map(lambda x: x +10, Nums))
print(result)
```

```python
c = [ 15,23,47]
result = list(map(lambda c:(c * 9/5) + 32, c))
print("Fahrenheit:", result)
```

```python
#Filter number > 10.
nums = [5,12,18,3,7]
result = list(filter(lambda x : x>10, nums))
print(result)
```

```python
words = ["Python", "", "data", "AI"]

result = list(filter(lambda x: x!= "", words))
print(result)
```

```python
#Data cleaning names from list
names = ["Alish", "Bob","Anna", "Charlie"]
name_filter = list(filter(lambda n : n.startswith("A"),names))
print(name_filter)
```

```python
#Positive number filter
n1 = [1,2,-1,-2,0,5,-7,8]
ptiv_num = list(filter(lambda x: x>0,n1))
print(ptiv_num)
```

```python
#Factorial
def factorial(n):
    if n ==0:
        return 1
    return n * factorial(n-1)
print(factorial(5))
```

```python
def sum_list(lst):
    if len(lst) == 0:
        return 0
    return lst[0] + sum_list(lst[1:])

print(sum_list([1, 2, 3, 4]))
```
