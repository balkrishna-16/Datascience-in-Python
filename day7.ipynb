```python
def login_system(user_role):
    print("Checking login creditials...")

    #Nested function: only accessible after login check
    def access_dashboard():
        print("Welcome to Dashboard")

    #Permission Check
    if user_role == "admin":
        print("Login succesful ")
        access_dashboard()     #inner function runs only if allowed

    else:
        print("Access denied ")

login_system("admin")
login_system("guest")
```

```python
def greet(name):
    return"Hello" + name
# Higher Order function
def process(func, value):
    return func(value)
print(process(greet," Sandesh"))
```

```python
square = lambda x: x*x         #x1: x1*x1
square(5)
```

```python
#Add two number using lamada
add = lambda x,y:x+y
add(3,2)
```

```python
#Use lamada to check if a number is even
is_even = lambda x: x % 2 == 0
is_odd = lambda x: x % 2 != 0

print(is_even(4))  # True
print(is_even(7))  # False
```

```python
#Get maximum of two numbers.
x_greater = lambda x, y : x>y 
y_greater = lambda x,y : x<y

print(x_greater(7,5))
print(y_greater(10,8))
```

```python
# minimum of two nums
x_lower = lambda x, y : x<y 
y_lower = lambda x,y : x>y

print(x_lower(7,5))
print(y_lower(10,8))
```

```python
#Get maximum of three numbers.
x_greater = lambda x, y,z : x>y |x>z
y_greater = lambda x,y,z : x<y  |y>z
z_greater = lambda x,y,z : x<y  |z >x


print(x_greater(7,5,9))
print(y_greater(10,8,6))
print(z_greater(10,8,6))
```

```python
#Return lenght of a string using lamada
x_length = lambda s: len(s)

print(x_length("hello"))  
print(x_length("Balkrishna"))  
```

```python
#map() applies a function to every item in a list
# Syntax:
# map(function, iterable)
num = [1,2,3,4]
result = list(map(lambda x: x*x, num))
print(num)
```

```python
#Conver list of strings to uppercase
words = ["Hello", "World", "Python"]
result = list(map(lambda w: w.upper(),words))
print(result)

#Lowercase from list of string
words = ["Hello", "World", "Python"]
result = list(map(lambda l: l.lower(),words))
print(result)
```

```python
#Add 10 to Each element in list
Nums = [1,2,3,4,5]
result = list(map(lambda x: x +10, Nums))
print(result)
```

```python
c = [ 15,23,47]
result = list(map(lambda c:(c * 9/5) + 32, c))
print("Fahrenheit:", result)
```

```python
# Condition filter
num = [1,2,3,4,5,6]
result= list(filter(lambda x: x %2 ==0, num))
print(result)
```
