```python
# Question:
# Project: Staff Management System (Core Version)

# Goal:
# Use list + dictionary + loop to manage staff records in a simple system.
# Features of the system 
#   -Add staff details 
#   -Store multiple staff members
#   -View all staff records
#   -Calculate salary summary
#   -Structured data using dictionary inside a main dicitionary

# # Staff Management System

staff_list = []

def add_staff():
    print("\n--- Add Staff ---")
    name = input("Enter name: ")
    age = int(input("Enter age: "))
    post = input("Enter post: ")
    salary = float(input("Enter salary: "))

    staff = {
        "name": name,
        "age": age,
        "post": post,
        "salary": salary
    }

    staff_list.append(staff)
    print("✅ Staff added successfully!")

    def view_staff():
        print("\n--- Staff Records ---")
    if len(staff_list) == 0:
        print("No staff data found!")
    else:
        for i, staff in enumerate(staff_list, start=1):
            print(f"\nStaff {i}")
            print("Name:", staff["name"])
            print("Age:", staff["age"])
            print("Post:", staff["post"])
            print("Salary:", staff["salary"])

def salary_summary():
    print("\n--- Salary Summary ---")
    if len(staff_list) == 0:
        print("No data available!")
    else:
        total = 0
        for staff in staff_list:
            total += staff["salary"]

        avg = total / len(staff_list)

        print("Total Salary:", total)
        print("Average Salary:", avg)
while True:
    print("\n===== Staff Management System =====")
    print("1. Add Staff")
    print("2. View Staff")
    print("3. Salary Summary")
    print("4. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        add_staff()
    elif choice == "2":
        view_staff()
    elif choice == "3":
        salary_summary()
    elif choice == "4":
        print("Exiting system...")
        break
    else:
        print("Invalid choice! Try again.")
```
