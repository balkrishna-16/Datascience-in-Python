```python
price = 500
discount_amount = 400

discount_percentage = discount_amount / price * 100

def calculate_discount(price, discount_percent):
    print("The price of monitor:", price)
    print("The discount percentage:", discount_percent, "%")

# function call 
calculate_discount(price, discount_percentage)
```

```python
c = 80
f = 80

def c_to_f(c):
    return (c * 9/5) + 32

def f_to_c(f):
    return (f - 32) * 5/9

# Function call and print
print("Celsius to Fahrenheit:", c_to_f(c))
print("Fahrenheit to Celsius:", f_to_c(f))
```

```python
nums = [1, 2, 3, 4, 5, 6, 7]

def calculate_sum(numbers):
    total = 0
    for num in numbers:
        total += num
    return total

result = calculate_sum(nums)
print("Sum is:", result)
```

# If i have function called:

```python
# If i have function called:
# add(1,2,3,4)

def add(*numbers):
    total = 0
    for num in numbers:
        total += num
    return total

result = add(1, 2, 3, 4)
print("Sum is:", result)
```

```python
def build_profile(**kwargs):
    return kwargs
user = build_profile(name="Ram", age=20, city="Kathmandu")
print(user)
```

```python
Data = {10, None, 30, "No", 40, 50}

def data_clean(data):
    clean_data = set()
    
    for item in data:
        if item is not None:
            clean_data.add(item)
    
    return clean_data

result = data_clean(Data)
print(result)
```
