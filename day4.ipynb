```python
students = []  # declaring a variable list to store values

n = int(input("Enter number of students (more than 5): "))

subjects = ["English", "Math", "Nepali", "Science"]

for i in range(n):
    name = input("Enter name: ")
    marks = []

    # Input marks using nested loop
    for sub in subjects:
        mark = int(input(f"Enter marks in {sub}: "))  # fixed
        marks.append(mark)  # append correct value

    total = sum(marks)
    percentage = total / len(subjects)

    # Fail condition
    if any(m < 32 for m in marks):
        status = "Fail"
        grade = "No Grade"
    else:
        status = "Pass"

        if percentage >= 80:
            grade = "A+"
        elif percentage >= 70:
            grade = "A"
        elif percentage >= 60:
            grade = "B"
        elif percentage >= 50:
            grade = "C"
        elif percentage >= 40:
            grade = "D"
        else:
            grade = "E"

    students.append({
        "name": name,
        "English": marks[0],
        "Math": marks[1],
        "Nepali": marks[2],
        "Science": marks[3],
        "total": total,
        "percentage": percentage,
        "status": status,
        "grade": grade
    })

print(students)

print("\n\n============================ STUDENT RESULT SHEET ================================\n")

for s in students:
    print("Name        :", s["name"]) 
    print("English     :", s["English"])
    print("Math        :", s["Math"])
    print("Nepali      :", s["Nepali"])
    print("Science     :", s["Science"])
    print("Total       :", s["total"])
    print("Percentage  :", round(s["percentage"], 2))
    print("Status      :", s["status"])

    if s["status"] == "Pass":
        print("Grade       :", s["grade"])
    else:
        print("Grade       : No Grade (Failed)")
    
    print("-" * 40)
```

```python
students = {
    "Ram": {"math": 80, "science": 70, "English": 90},
    "Shyam": {"math": 90, "science": 85, "English": 88},
    "Hari": {"math": 60, "science": 65, "English": 70}
}

for name, subjects in students.items():
    total = sum(subjects.values())
    avg = total / len(subjects)

    if avg >= 80:
        grade = "A"
    elif avg >= 60:
        grade = "B"
    else:
        grade = "C"

    print(f"{name}: Average: {avg:.2f}, Grade: {grade}")
```

```python
users = {
    "anshu": "1234",
    "bkc": "abcd"
}

attempts = 3

while attempts > 0:
    username = input("username: ")
    password = input("password: ")

    if username in users and users[username] == password:
        print("\033[92mLogin successful\033[0m")
        break
    else:
        attempts -= 1
        print("\033[91mWrong! Attempts left:", attempts, "\033[0m")

if attempts == 0:
    print("\033[91mAccount locked!\033[0m")
```

```python
cart = {
    "apple":{"price":100, "qty":2},
    "banana":{"price":50, "qty":5},
    "milk":{"price":80, "qty":1},
}
total = 0
for item,details in cart.items():
    cost = details["price"] * details["qty"]
    total += cost
    print(item, ":", cost)

print("Total bill: ", total)
```

```python
menu = {
    "momo": {"price": 100},
    "pizza": {"price": 250},
    "biryani": {"price": 380}
}

print("\n=========== BILL ===========")
total = 0

for food, details in menu.items():
    cost = details["price"]
    total += cost
    print(f"{food.capitalize()} : Rs {cost}")

print("-----------------------------")
print(f"Total Bill: Rs {total}")
```

```python
customer = input("Enter customer name: ")

cart = []
grand_total = 0

while True:
    name = input("Enter product name: ")

    # input validation
    try:
        price = int(input("Enter price: "))
        qty = int(input("Enter quantity: "))
    except ValueError:
        print("Invalid input! Please enter numbers only.")
        continue

    total = price * qty
    grand_total += total

    cart.append({
        "name": name,
        "price": price,
        "qty": qty,
        "total": total
    })

    choice = input("Add more items? (yes/no): ").lower()
    if choice == "no":
        break

print("\n=================== BILL ===================")
print("Customer:", customer)
print("-" * 40)

for i, item in enumerate(cart, start=1):
    print(f"{i}. {item['name']} Price: {item['price']} Qty: {item['qty']} Total: {item['total']}")

print("-" * 40)
print("Grand Total:", grand_total)
```
